````python
from skimage import io, color
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
import numpy as np
print("All Packages Loaded")
````

**Load Data**

````python
image_path = '../input/cars-image-dataset/Cars Dataset/train/Hyundai Creta/122.jpg'

# Load the image into a NumPy array
image = io.imread(image_path)
````

````python
print("-----------------------------------------")
print("Type of image - ",type(image))
print("-----------------------------------------")
print("Image Shape ",image.shape)
print("-----------------------------------------")
print("Image Size ",image.size)
print("-----------------------------------------")
````

## Image Show Function

````python
def image_show(image, title='Image'):
    plt.imshow(image)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()


    

def image_show_two(original, filtered_image, filtered_title):
    fig,(ax1, ax2)=plt.subplots(ncols= 2, figsize=(8,6))
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original Image')
    ax1.axis('off')
    ax2.imshow(filtered_image, cmap=plt.cm.gray)
    ax2.set_title(filtered_title)
    ax2.axis('off')
````

````python
image_show(image)
image.shape
````

**Flipping images: vertically**

````python
# Flip the image vertically
seville_vertical_flip = np.flipud(image)
image_show(seville_vertical_flip)
````

**Flipping images: horizontally**

````python
# Flip the previous image horizontally
seville_horizontal_flip = np.fliplr(image)
image_show(seville_horizontal_flip)
````

## Visualizing histograms with Matplotlib



````python
# Obtaining the red values of the image
red_channel = image[:, :, 0]
# Obtaining the green values of the image
green_channel = image[:, :, 1]
# Obtaining the blue values of the image
blue_channel = image[:, :, 2]



# Plot the histogram for the red channel
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.hist(red_channel.ravel(), bins=256, color='red')
plt.title('Red Channel')

# Plot the histogram for the green channel
plt.subplot(132)
plt.hist(green_channel.ravel(), bins=256, color='green')
plt.title('Green Channel')

# Plot the histogram for the blue channel
plt.subplot(133)
plt.hist(blue_channel.ravel(), color='blue', bins=256)
plt.title('Blue Channel')

# Show the plots
plt.show()
````

````python
# Create a 2x2 subplot grid
fig = plt.figure(figsize=(10, 10))

# Place the original image at the center of the top row
ax1 = plt.subplot2grid((3, 3), (0, 1), colspan=1)
ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')

# Plot the red channel
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=1)
ax2.imshow(red_channel, cmap='gray')
ax2.set_title('Red Channel')
ax2.axis('off')

# Plot the green channel
ax3 = plt.subplot2grid((3, 3), (1, 1), colspan=1)
ax3.imshow(green_channel, cmap='gray')
ax3.set_title('Green Channel')
ax3.axis('off')

# Plot the blue channel
ax4 = plt.subplot2grid((3, 3), (1, 2), colspan=1)
ax4.imshow(blue_channel, cmap='gray')
ax4.set_title('Blue Channel')
ax4.axis('off')



# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
````

## Thresholding

**Only from grayscale images**

Simplest method of image segmentation

By making it black and white


- 255 (white) if pixel > thresh value
- 0 (black) if pixel < thresh value


## Global or histogram based: good for uniform backgrounds


````python
# Import the otsu threshold function
from skimage.filters import threshold_otsu

# Make the image grayscale using rgb2gray
image_gray = rgb2gray(image)

# Obtain the optimal threshold value with otsu
thresh = threshold_otsu(image_gray)

# Apply thresholding to the image
binary = image_gray > thresh

# Show the image
image_show_two(image, binary, "Thresholing Image")
````

**When you are not being sure about what thresholding method to use isn't a problem.  It returns a figure comparing the outputs of different global thresholding methods. try_all_threshold() to try multiple global algorithms.**

````python
# Import the try all function
from skimage.filters import try_all_threshold

# Use the try all method on the resulting grayscale image
fig, ax = try_all_threshold(image_gray, verbose=False)

# Show the resulting plots
plt.show()
````
